## Last Digits of Powers

```python
import networkx as nx
import matplotlib.pyplot as plt
from math import cos, sin, pi
```

```python
def plot_digit_trace(n, factor, rad=10):
    plt.gca().set_aspect('equal')
    graph = nx.DiGraph()
    graph.add_edges_from([(i, i * factor % n) for i in range(n)])

    positions = [(rad * cos(i * 2 * pi / n), (rad * sin(i * 2 * pi / n)))
            for i in range(n)]

    nx.draw(graph, pos=positions, with_labels=True,
            font_color='white', node_size=400, font_size=12)
```

```python
plot_digit_trace(10, 2)
```

```python
plot_digit_trace(10, 3)
```

```python
plot_digit_trace(10, 5)
```
